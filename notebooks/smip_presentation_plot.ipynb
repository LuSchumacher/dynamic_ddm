```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FONT_SIZE_1 = 18
FONT_SIZE_2 = 16
FONT_SIZE_3 = 14
```

```python
f, ax = plt.subplots(2, 2, figsize=(12, 6))

# classic approach
y = np.repeat(1, 800)
x = np.arange(start=1, stop=801)
sd = 0.2

ax[0, 0].plot(x, y, color='maroon', lw=2)
ax[0, 0].fill_between(x, y-sd, y+sd, color='maroon', alpha=0.5, linewidth=0)
ax[0, 0].set_ylabel('Parameter\nValue', fontsize=FONT_SIZE_2, rotation=0, labelpad=60)
ax[0, 0].set_xlabel('\nTime', fontsize=FONT_SIZE_2)
ax[0, 0].set_ylim(0, 2)
ax[0, 0].tick_params(axis='both', which='major', labelsize=FONT_SIZE_3)
ax[0, 0].set_title("Classical approach", size=FONT_SIZE_1)
ax[0, 0].set_xticks([])
ax[0, 0].set_yticks([])

# random variablity
y = np.random.normal(1, 0.2, 800)
x = np.arange(start=1, stop=801)
sd = 0.2

ax[0, 1].plot(x, y, color='maroon', lw=1, alpha=0.8)
ax[0, 1].fill_between(x, y-sd, y+sd, color='maroon', alpha=0.1, linewidth=0)
ax[0, 1].set_ylim(0, 2)
ax[0, 1].tick_params(axis='both', which='major', labelsize=FONT_SIZE_3)
ax[0, 1].set_title("Stationary variability", size=FONT_SIZE_1)
ax[0, 1].set_xticks([])
ax[0, 1].set_yticks([])


# deterministic functions
y1 = 1.5 - 1/800 * np.arange(800)
y2 = 2.0 * (1 - 0.005)**np.arange(800)
x = np.arange(start=1, stop=801)
sd = 0.2

ax[1, 0].plot(x, y1, color='maroon', lw=2)
ax[1, 0].fill_between(x, y1-sd, y1+sd, color='maroon', alpha=0.5, linewidth=0)
ax[1, 0].plot(x, y2, color='green', lw=2)
ax[1, 0].fill_between(x, y2-sd, y2+sd, color='green', alpha=0.5, linewidth=0)
ax[1, 0].set_ylim(0, 2)
ax[1, 0].tick_params(axis='both', which='major', labelsize=FONT_SIZE_3)
ax[1, 0].set_title("Deterministic function", size=FONT_SIZE_1)
ax[1, 0].set_xticks([])
ax[1, 0].set_yticks([])

# superstatistic
y = np.zeros(800)
y[0] = 1
for i in range(1, 800):
    y[i] = min(max(y[i - 1] + 0.05 * np.random.normal(0, 1), 0.0), 2.0)

ax[1, 1].plot(x, y, color='maroon', lw=1, alpha=0.8)
ax[1, 1].fill_between(x, y-sd, y+sd, color='maroon', alpha=0.5, linewidth=0)
ax[1, 1].set_ylim(0, 2)
ax[1, 1].tick_params(axis='both', which='major', labelsize=FONT_SIZE_3)
ax[1, 1].set_title("Non-stationary variability", size=FONT_SIZE_1)
ax[1, 1].set_xlabel('\nTime', fontsize=FONT_SIZE_2)
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])

f.tight_layout()
sns.despine()
f.savefig("../plots/talk_graphic.png", dpi=300, bbox_inches="tight")
```
